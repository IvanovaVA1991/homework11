```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a, b, c, d, e = -12, -18, 5, 10, -30
start = -20*np.pi
stop = 20*np.pi
x = np.arange(start, stop, 0.01)
def func(x):
    function = a*x**4*np.sin(np.cos(x))+b*x**3+c*x**2+d*x+e
    return function
```

```python
def graph(x):
    plt.title('График функции')
    plt.xlabel('Ось X')
    plt.ylabel('Ось Y')
    plt.grid()
    plt.plot(x, func(x), 'g')
    plt.show()
    print(('Функция синусоидная, поэтому корней, интервалов возрастания и убывания функции бесконечное множество'))
graph(x)
```

```python
def rootsCount():
    global start
    global stop
    count = 0
    x_temp = func(start)
    for x in np.arange(start, stop, 0.01):
        if x_temp * func(x) < 0: count+=1
        x_temp = func(x)
    print(f'На выбранном интервале {count} корней, вершин, интервалов возрастания и убывания функции')
rootsCount()
```
